<h2 style="color:rgb(0,120,170)">Task 1 (20 Points): Prove the following Lemma</h2>

Show that for any matrix $X$ and any differentiable scalar-valued function $f$ it holds that 
$$
\frac{\partial f(X^\top X)}{\partial X} = 2{X}\frac{\partial f(X^\top X)}{\partial (X^\top X)}.
$$

########## YOUR SOLUTION HERE ##########


Using the chain rule we have:
\begin{equation}
\frac{\partial f(X^\top X)}{\partial X}= \frac{\partial f(X)}{\partial X^\top X}\frac{\partial X^\top X}{\partial X}=  \frac{\partial f(X)}{\partial X^\top X}\cdot 2X = 2{X}\frac{\partial f(X^\top X)}{\partial (X^\top X)}
\end{equation}

$$
\nabla_{\mathbf{U}} \ell(\mathbf{U}, \mathbf{\Psi}) = 2 \mathbf{P}\mathbf{U}
$$
and
$$
\nabla_{\mathbf{\Psi}} \ell(\mathbf{U}, \mathbf{\Psi}) = \operatorname{diag}(\mathbf{P})
$$

$$
\mathbf{P} = \nabla_{\mathbf{U}\mathbf{U}^\top + \mathbf{\Psi}} \ell(\mathbf{U}, \mathbf{\Psi}) = \frac12 \mathbf{Q} - \frac{1}{2n} \mathbf{Q} \mathbf{X}^\top \mathbf{X} \mathbf{Q}
$$
and
$$
\mathbf{Q} = (\mathbf{U}\mathbf{U}^\top + \mathbf{\Psi})^{-1}.
$$ 

Hint: use the fact that if $f(\mathbf{A})=f(\mathbf{A}^T)$ holds for a scalar function $f$ and a real square matrix $\mathbf{A}$ it holds that $\nabla_{\mathbf{X}} f(\mathbf{X}^\top \mathbf{X}) = 2 \mathbf{X} \nabla_{\mathbf{X}^\top \mathbf{X}} f(\mathbf{X}^\top \mathbf{X})$. 

########## YOUR SOLUTION HERE ##########

Since $x \sim \cal{N} (0, \mathbf{U}\mathbf{U}^\top + \mathbf{\Psi})$:

\begin{align}
\ell(\mathbf{U}, \mathbf{\Psi}) &= -\log\mathcal{L}(\mathbf{X}; \mathbf{U}, \mathbf{\Psi})\\
&= \frac{n}{2} \log \det(\mathbf{UU}^\top + \mathbf{\Psi}) + \frac{1}{2} \sum_{i=1}^n \mathbf{x}_i^\top (\mathbf{UU}^\top + \mathbf{\Psi})^{-1} \mathbf{x}_i \\
&= \frac{n}{2} \log \det(\mathbf{Q}^{-1}) + \frac{1}{2} \sum_{i=1}^n \mathbf{x}_i^\top \mathbf{Q} \mathbf{x}_i
\end{align}


\begin{align}
\nabla_{\mathbf{U}} \ell(\mathbf{U}, \mathbf{\Psi}) &= \frac{\partial}{\partial \mathbf{U}} 
\left( \frac{n}{2} \log \det(\mathbf{UU}^\top + \mathbf{\Psi}) 
+ \frac{1}{2} \sum_{i=1}^n \mathbf{x}_i^\top (\mathbf{UU}^\top + \mathbf{\Psi})^{-1} \mathbf{x}_i \right)\\
\text{by chain rule} \qquad
&= \frac{n}{2} \frac{\partial \log \det (\mathbf{UU}^\top \mathbf{\Psi})}{\partial \mathbf{(\mathbf{UU}^\top + \mathbf{\Psi})}}

\text{by eq (57), (61), (62)} \qquad

&= 2 n \left(\frac{1}{2}\mathbf{Q} - \frac{1}{2n} \mathbf{QX}^\top \mathbf{X}\mathbf{Q} \right) \mathbf{U}\\
&= 2 \mathbf{PU} \qquad \text{after normalization}
\end{align}

\begin{align}
\nabla_{\mathbf{\Psi}} \ell(\mathbf{U}, \mathbf{\Psi}) &= \frac{\partial}{\partial \mathbf{\Psi}}

&= \frac{n}{2} (\mathbf{UU}^\top + \mathbf{\Psi})^{-1}\cdot 1
+ \frac{1}{2}\sum_{i=1}^n -\mathbf{Qx}_i^\top \mathbf{x}_i\mathbf{Q} \cdot \mathbf{I} \cdot 1\\
&= n\frac12 \mathbf{Q} - \frac12 \mathbf{Q} \mathbf{X}^\top \mathbf{X} \mathbf{Q} = \frac12 \mathbf{Q} - \frac{1}{2n} \mathbf{Q} \mathbf{X}^\top \mathbf{X} \mathbf{Q}

<h2 style="color:rgb(0,120,170)">Task 3 (20 Points): Generate toy data for FA</h2>

Generate an $n \times m$ data matrix $\mathbf{X}$, i.e. $n=100$ samples with $m=5$ features, according to the generative factor analysis model with $l=3$ factors. Further, fill $\mathbf{U}$ with random integers from the set $\{-3, \dots, 3\}$ and $\mathbf{\Psi}$ with random integers from the set $\{1, \dots, 3\}$. To save memory, implement $\mathbf{\Psi}$ as a vector. 

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math, Latex

########## YOUR SOLUTION HERE ##########
def generate_data(seed):
    np.random.seed(seed)
    
    n = 100  
    m = 5    
    l = 3    
    
    U = np.random.randint(-3, 4, size=(m, l))
    
    Psi = np.random.randint(1, 4, size=m)
    
    Y = np.random.randn(n, l)
    
    eps = np.random.randn(n, m) * np.sqrt(Psi)
    
    X = Y @ U.T + eps

    return X    

print(generate_data(0)[:5])
```

<h2 style="color:rgb(0,120,170)">Task 4 (20 Points): Implement GD for FA</h2>

Use your results from the previous task to implement gradient descent on the negative log-likelihood for factor analysis. Initialize $\mathbf{U}$ and $\mathbf{\Psi}$ randomly uniform in $[-0.1, 0.1]$. To speed up computations, use the matrix inversion lemma $(\mathbf{UU}^\top + \mathbf{\Psi})^{-1} = \mathbf{\Psi}^{-1} - \mathbf{\Psi}^{-1}\mathbf{U}(\mathbf{I} + \mathbf{U^\top \mathbf{\Psi}^{-1} \mathbf{U}})^{-1}\mathbf{U}^\top \mathbf{\Psi}^{-1}$ wherever possible and adjust the learning rate properly. Visualize the learning progress over 100 update steps in terms of the loss function $\ell(\mathbf{U},\mathbf{\Psi})$ for 10 runs which differ only by initialization.

```python
########## YOUR SOLUTION HERE ##########

def Q_(U, Psi):
    Psi_inv= 1/Psi # (m)
    Psi_inv_U = Psi_inv[:, None] * U  # (m,l)
    I = np.eye(U.shape[1]) # (m,l)
    U_T_Psi_inv = Psi_inv_U.T # (l,m)
    U_T_Psi_inv_U = U.T @ Psi_inv_U # (l,l)

    Q = np.diag(Psi_inv) - Psi_inv_U @ np.linalg.inv(I+U_T_Psi_inv_U)@U_T_Psi_inv # (m, m)
    return Q
    
def loss(X, Q):
    n = X.shape[0]
    
    _, logdet_Q = np.linalg.slogdet(Q)
    sum_xT_Q_x = sum(x_i.T @ Q @ x_i for x_i in X)
    
    loss = 0.5 * (-n*logdet_Q + sum_xT_Q_x)
    return loss

def gd(X, seed,l=3, lr=1e-4, steps=100):
    n, m = X.shape
    np.random.seed(seed)
    U = np.random.uniform(-0.1, 0.1, size=(m, l))
    Psi = np.random.uniform(0.01, 0.1, size=m) 
    losses = []

    for step in range(steps):    
        Q = Q_(U, Psi)
        
        P = 0.5 * Q - 0.5 * Q @ X.T @ X @ Q /n
        
        U -= lr * 2 * P @ U
        Psi -= lr * np.diag(P)

        Q_upd = Q_(U, Psi)
        
        loss_ = loss(X, Q_upd)
        losses.append(loss_)
    return losses

def normalize(X):
    mean = np.mean(X, axis=0)
    std = np.std(X, axis=0)
    return (X - mean) / std
  
runs = 10
losses = []
for i in range(runs):
    X = generate_data(i)
    X=normalize(X)
    losses_X = gd(X, i)
    losses.append(losses_X)

plt.figure(figsize=(10, 6))
for i in range(runs):
    plt.plot(np.array(losses[i]) - losses[i][-1])
    # plt.plot(losses[i], label=f'Run {i+1}')
    
plt.xlabel("Step")
plt.ylabel("Loss")
plt.title("Learning progress: GD")
plt.show()
```

<h2 style="color:rgb(0,120,170)">Task 5 (20 Points): Implement EM for FA</h2>

Implement expectation maximization for factor analysis. Use the same data and initialization scheme as in the previous task. Visualize the learning progress over 100 update steps in terms of the loss function $\ell(\mathbf{U},\mathbf{\Psi})$ for 10 runs which differ only by initialization. Compare your results to those in the previous exercise and interpret the ovserved differences. 

Use Speedups from lecture notes!

```python
########## YOUR SOLUTION HERE ##########

def C_(X):
    n,_= X.shape
    return (X.T@X)/n 
    
def U_hat_(Psi,U):
    Psi_inv = 1/Psi
    return Psi_inv[:, None] * U 
    
def B_(U,U_hat):
    return U.T@U_hat

def E_step(Q, U, Psi, X, I_l):
    #Ey|x[log p(x, y)]
    n, m = X.shape
    l = U.shape[1]

    E_y = (U.T @ Q @ X.T).T # (n, l) 

    Sigma = I_l - U.T @ Q @ U # (l, l) 

    E_yyT = np.empty((n, l, l)) 
    for i in range(n):
        mu = E_y[i, :, None] # (l, 1)
        E_yyT[i] = mu @ mu.T + Sigma 

    return E_y, E_yyT

def M_step(X,U, Psi, E_, C, I):
    # U , Ψ = arg maxU ,Ψ Ey|x[log p(x, y)]
    
    n, m = X.shape
    E_y, E_yyT = E_ 
    
    U_hat = U_hat_(Psi,U)
    B=B_(U,U_hat)

    avg_xEy = C@(U_hat - U_hat@np.linalg.inv(I+B)@B) # speedup 5.42
    
    avg_EyyT = np.mean(E_yyT, axis=0)  # (l, l)
    U_new = avg_xEy@np.linalg.inv(avg_EyyT) # 5.39

    Psi_new = np.diag(C - (X.T @ E_y @ U_new.T)/n) # 5.40 
    
    return U_new, Psi_new

def EM(X, C, seed, l=3, lr=1e-4, steps=100):
    n, m = X.shape
    np.random.seed(seed)
    U = np.random.uniform(-0.1, 0.1, size=(m, l))
    Psi = np.random.uniform(0.01, 0.1, size=m) 
    I_l = np.identity(l)
    losses = []

    for step in range(steps):    
        Q = Q_(U, Psi)

        E_ = E_step(Q, U, Psi, X, I_l)
        
        U, Psi = M_step(X, U, Psi, E_, C, I_l)

        Q_new = Q_(U, Psi)
        
        loss_ = loss(X, Q_new)
        losses.append(loss_)
    return losses

# print(Q.shape, U.shape, Psi.shape)
#(5, 5) (5, 3) (5,)

runs = 10
losses = []
for i in range(runs):
    X = generate_data(i)
    X=normalize(X)
    C = C_(X)
    losses_X = EM(X, C, i)
    losses.append(losses_X)

plt.figure(figsize=(10, 6))
for i in range(runs):
    plt.plot(np.array(losses[i]) - losses[i][-1])

plt.xlabel("Step")
plt.ylabel("Loss")
plt.title("Learning progress: EM")
plt.show()
```
